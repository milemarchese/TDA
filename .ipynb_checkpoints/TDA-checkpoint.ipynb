## TP1 Gale-Shapley

#### Parte 1: El club "PICA-PICA"

***Gale-Shapley con empates.***

Vamos a resolver Gale-Shapley con indiferencia entre candidatos resolviendo empates de dos formas: orden alfabético y random. Podemos considerar que estamos usando la variante weakly-stable ya que estamos rompiendo los empates de forma arbitraria y después aplicamos Gale-Shapley simple para obtener las parejas. **Siempre hay match. No es asi en super y strongly-stable.**

Podemos probar que es estable ya que si hay igual cantidad de oferentes que candidatos si hay un oferente que puede hacer propuestas entonces va a haber un candidato libre.

Preliminarmente, el Gale-Shapley implementado tiene la siguiente invariante: si un jugador j del grupo de jugadores que aceptan/rechazan propuestas todavía no tiene pareja entonces acepta el primer ofrecimiento y luego, si es que se lo proponen, puede llegar a cambiar a su pareja por otra ques esté mejor posicionada en su ranking a medida que avanza el algoritmo, es decir, **el jugador j nunca deja de estar comprometido con alguien desde que tiene su primer propuesta**. No es así respecto a los jugadores que hacen las propuestas, estos pueden ir cambiando su estado durante la ejecución ya que pueden ser rechazados si su pareja consigue una mejor, esto quiere decir que el ranking de su pareja va decayendo a medida que hace propuestas.

***Observaciones del Tardos***

1.

m prefers w to w′.
m′ prefers w to w′.
w prefers m to m′.
w′ prefers m to m′.

Here is a unique stable matching here, consisting of the pairs(m,w) and (m′,w′). The other perfect matching, consisting of thepairs(m′,w) and (m,w′), would not be a stable matching, because the pair(m,w)would form an instability with respect to this matching.

2.

m prefers w to w′.
m′ prefers w′ to w.
w prefers m′ to m.
w′ prefers m to m′.

In  this  second  example,  there  are  two  different  stable  matchings.  The matching consisting of the pairs(m,w) and (m′,w′)is stable, because both men are as happy as possible, so neither would leave their matched partner.But the matching consisting of the pairs(m′,w) and (m,w′)is also stable, for the complementary reason that both women are as happy as possible. It’s possible for an instance to have more than one stable matching. *El matching resultante depende de quienes son los oferentes que siempre es el más "feliz".*


***Complejidad***

Para armar las listas de preferencias se recorren los n jugadores para obtener sus datos, y a su vez por cada jugador se tiene recorrer las listas de preferencias de cada uno de ellos, que contienen a su vez n jugadores, por ende, cargar las listas de preferencias toma O(n2).
En el peor caso posible cada jugador oferente va a preguntarle una vez a cada uno de sus candidatos, eso resulta O(n2). Todas las operaciones realizadas como preguntar si está libre, reemplazar una pareja por otra, etc. son operaciones O(1).

#### Parte 2: Funciones matemáticas.

Operaciones con listas, vectores.

```python
def gale_shapley(men: list):
    pairs = set()
    while len(men):
        man = men.pop()
        woman = man.pop_preferred()
        if woman.is_free():
            woman.engage(man)
            man.engage(woman)
            pairs.add((man, woman))
        else:
            if woman.prefers(man):
                ex = woman.release()
                pairs.remove((ex, woman))
                ex.release()
                men.append(ex)
                woman.engage(man)
                man.engage(woman)
                pairs.add((man, woman))

        if man.can_propose():
            men.append(man)
    return pairs
```

### Parte 2 -  Programando la producción semanal.

Una empresa que produce partes de aviones solicita nuestros servicios para planificar su producción de las próximas n semanas. Produce 3 tipos de piezas complejas. Cada pieza lleva una semana de elaboración. Por restricciones espaciales solo puede generar un tipo de pieza simultáneamente. Asimismo cada vez que produce un tipo de pieza, no puede volver a realizarla en la siguiente semana para evitar el sobrecalentamiento de cierta maquinaria.

***Greedy***

Suponiendo que la planilla de pedidos no tiene ofertas repetidas en la semana y que tenemos una pequeña API que, siendo pi una pieza, ki el precio ofrecido por la pieza pi y Si la semana i nos entrega una colección A de Si = {(p1, k1), (p2, k2), (p3, k3)} con i = 1, 2 , 3.
Ordenamos los productos de cada semana de mayor a menor y elegimos la semana que da mas ganancia. Contemplamos que no se puede producir lo mismo dos semanas seguidas. Optimizamos empates.

***Análisis de complejidad***

Vamos a ordenar cada semana de mayor a menor en Θ(n log n). Luego siendo n la cantidad de ofertas, recorremos cada semana hasta que encontramos el que tenga el mayor primer elemento, en el peor de los casos hay un empate de ofertas en la semana y entonces buscamos el siguiente mayor elemento, estos ciclos anidados son Θ(n2), la verificación para detectar los empates que hacen que el ciclo no termina es O(1) ya que es un set de 3 elementos para recorrer. Luego en Θ(1) agregamos la semana óptima a nuestra solución y tenemos en cuenta que la máquina no puede trabajar dos semanas seguidas, entonces recorremos todos los elementos de todas las semanas y borramos en Θ(1) el que no cumpla con esta condición. En total el último ciclo es Θ(n) y considerando el peor de los casos nuestro greedy es Θ(n2).
Podríamos optimizar el algoritmo si contamos con una estructura, como un heap de máximos que bajo el criterio es el precio ofrecido por el primer elemento de la semana (previamente ordenada de mayor a menor). 

***Programación dinámica***

Siendo A una matriz de 3x(n+1), donde n es la cantidad de semanas, la primer columna se inicializa con ceros y las siguientes con el ofrecimiento monetario de cada tipo de pieza.

Relaciones de recurrencia:

El problema requiere de una matriz auxiliar para almacenar las ganancias totales hasta la semana i. La relación de recurrencia para obtener esta matriz es la siguiente:

ganancias[0][j] = 0
si i 0 ganancias[i][j] = A[i][j] + maxgananciasi - 1[e]  / e[1, 3] e  j
donde A es la matriz de datos y ganancias[i][j]se refiere al elemento de la matriz ganancias ubicado en la columna i, fila j.
El vector de resultados se obtiene a partir de la matriz de ganancias de la siguiente forma:
R[0] = argmax(e)gananciasi - 1[e]  / e[1, 3] 
si i  n Ri = argmax(e)gananciasi - 1[e]  / e[1, 3] e  R[i + i]

***Análisis de complejidad***

Para un análisis general, suponemos que se pueden fabricar k tipos de piezas.
La función max_distinto(col, elem) recorre los elementos de una columna, por lo que su complejidad es O(k).
La función maximizar_ganancia(A) recorre primero todos los elementos de la matriz (n*k), y para cada uno realiza una asignación que utiliza a max_distinto, lo que es de orden O(k2n). Luego, recorre un vector de resultados de longitud n, utilizando también la función max_distinto para asignar un valor. Por lo tanto, esta parte es O(kn).
Por lo tanto el orden total del algoritmo es O(k2n + kn), que se puede aproximar por O(k2n).

### Parte 3 - Clases de complejidad

***P***

Sea un problema X: El algoritmo A soluciona x si para todo x ∈ X retorna una solución Sol(A) que es la solución óptima del problema. A x lo podemos representar como una secuencia de bits de longitud |x|. Si existe una solución p polinómica:
A = Θ(p(X)) 
Entonces X ∈ P. P suele ser la clase de problemas computacionales que son “eficientemente resolubles” o “tratables”. Existen algoritmos "galácticos".

***NP***

El algoritmo B certifica a x si para todo x ∈ X B(x, c) devuelve si si Sol(x) = c y no en caso contrario. Si existe una solución p polinómica, p dependiendo de |x|, |c|, B = Θ(p(x)) entonces X ∈ NP, quiere decir que pueden ser verificados en tiempo polinomial. Podemos decir que los problemas NP son problemas de decisión resueltos en tiempo polinomial de forma no determinística.

***NP-hard***

Un problema X es NP-hard si para todo problema Y ∈ NP se puede reducir a X. Entonce NP-hard contiene a los problemas de decisión que son como mínimo tan difíciles como un problema de NP. 

***NP-complete***

si X ∈ NP y  X es NP-hard. Los problemas NP-completos se pueden transformar unos en otros por una reducción polinómica y de tenerse una solución polinómica para un problema NP-completo, todos los problemas de NP tendrían también una solución en tiempo polinómico.

***Demostrar que es NP complete:***

* Probar que existe un verificador.
* Reducir un NP-completo a nuestro problema.
